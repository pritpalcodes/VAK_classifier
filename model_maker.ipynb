```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python

# Assuming the DataFrame is already loaded
df = pd.read_csv('synthetic_learning_data.csv')

# Generate new column names as 'column1', 'column2', 'column3', ...
new_column_names = [f'column{i+1}' for i in range(len(df.columns))]

# Rename the columns
df.columns = new_column_names

# Save the updated DataFrame to a new CSV file (optional)
df.to_csv('renamed_columns_data.csv', index=False)

# Display the first few rows of the DataFrame with renamed columns
print(df.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
y = df["column1"]
x=df.iloc[:, 1:]

# print(x.shape)
# print(y.shape)
x.sample(5)
```

```python
# Label encode the target variable Y
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_encoded
```

```python
# Assuming all categorical features need to be OneHotEncoded
categorical_features = x.select_dtypes(include=['object']).columns

# Apply OneHotEncoding to categorical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'
)

x_processed = preprocessor.fit_transform(x)
```

```python
x_processed[5]
```

```python
#split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x_processed, y_encoded, test_size=0.2, random_state=42)
```

```python
#Build and train the Multi-Class classification using RandomForest
multi_class_model = RandomForestClassifier(random_state=42)
multi_class_model.fit(x_train, y_train)
```

```python
# Predictions and evaluation
y_pred = multi_class_model.predict(x_test)
print("Multi-Class Classification Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

### Multi-label Classification 

```python
# OneHotEncode the target variable Y for multi-label classification
y_multi_label = pd.get_dummies(y)
```

```python
x_train, x_test, y_train_multi, y_test_multi = train_test_split(x_processed, y_multi_label, test_size=0.2, random_state=42)
```

```python
# Multi-Label classification using RandomForest
multi_label_model = MultiOutputClassifier(RandomForestClassifier(random_state=42))
multi_label_model.fit(x_train, y_train_multi)
```

```python
# Predictions and evaluation
y_pred_multi = multi_label_model.predict(x_test)
print("Multi-Label Classification Accuracy:", accuracy_score(y_test_multi, y_pred_multi))
print("\nClassification Report:\n", classification_report(y_test_multi, y_pred_multi, target_names=y_multi_label.columns))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
import joblib

# Assuming `df` is your original DataFrame
x = df.iloc[:, 1:]  # Features
y = df["column1"]    # Target

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Identify categorical features in X
categorical_features = x.select_dtypes(include=['object']).columns

# Preprocessor to handle categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'
)

# Create a pipeline that combines preprocessing and the model
multi_class_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Train the pipeline on the original DataFrame (no train/test split)
multi_class_pipeline.fit(x, y_encoded)

# Save the pipeline to a .pkl file
joblib.dump(multi_class_pipeline, 'multi_class_model.pkl')
```
